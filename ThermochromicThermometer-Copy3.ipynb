```python
###
### Author: Sam Ferrante
### Thermochromism: This program will produce a recipe for a solution or set of solutions that will change color from pink to blue at a temperature specified by the user
###
### Description: This program will first ask for a number of solutions the user would like to make (any positive # =< 50), 
###              next it will ask for the volume of each solution and if the user provided temperatures are in Fahrenheit, Celcius, or Kelvin, F/C/K, 
###              If the number of the solutions is 1 then it will proceed to ask for a single temperature and then return the recipe, 
###              otherwise it will ask if the temperatures are in a series (every 1, 2, 5... etc degrees) Yes/No,
###              if 'yes' it will ask for a starting temperature, and the difference between each marked temp. 
###              if 'no' it will ask for each specific temperature corresponding to the total number of solutions indicated in the first question.
###              finally it will spit out a table containing the recipe required for each solution.

import numpy as np

list_of_IPA = []
list_of_MEOH = []
temps = []
                
def main():
    num_solns = ''
    while num_solns.isnumeric() == False:
        num_solns = (input('Enter the number of solutions you wish to produce (must be less than or equal to 50): \n'))
    while int(num_solns) > 50 or int(num_solns)< 1:
        num_solns = ''
        while num_solns.isnumeric() == False: 
            num_solns = (input('Response must be an integer greater than 0 or less than 51: \n'))
    num_solns = int(num_solns)

    
    vol_solns = ''
    while vol_solns.isnumeric() == False:
        vol_solns = (input('Enter total volume of the solutions you wish to produce (in mL): \n'))
    while int(vol_solns) < 1: 
        vol_solns = ''
        while vol_solns.isnumeric() == False: 
            vol_solns = (input('Volume must be greater than 0: \n'))
    vol_solns = int(vol_solns)

    mass = (.4/100) * (vol_solns)
    
    metric = input('Enter the temperature scale from the choices; Fahrenheit, Celsius, or Kelvin (F/C/K): \n')
    while metric != 'F' and metric != 'C' and metric != 'K' and metric != 'f' and metric != 'c' and metric != 'k':
        metric = input('Must answer either (F, C, or K): \n')
    
    if metric == 'f' or metric == 'F':
        metric = 'Fahrenheit'
    elif metric == 'c' or metric == 'C':
        metric = 'Celsius'    
    elif metric == 'k' or metric == 'K':
        metric = 'Kelvin'
    
    series_yesno =''
    
    if num_solns != 1:
        series_yesno = str(input('Is this a series of temperatures with a constant value separating each temperature? (Yes/No) \n'))
        while series_yesno!= 'Yes' and series_yesno != 'yes' and series_yesno != 'YES' and series_yesno != 'Y' and series_yesno != 'y' and series_yesno != 'NO' and series_yesno != 'No' and series_yesno != 'no' and series_yesno != 'n' and series_yesno != 'N':
            series_yesno = input('Must answer either (Yes or No): \n')
        
    if series_yesno == 'Yes' or series_yesno == 'yes' or series_yesno == 'YES' or series_yesno == 'Y' or series_yesno == 'y':
        it_is = False
        while it_is == False:
            series = (input('What is the constant value between each temperature? \n'))
            try:
                float(series)
                it_is = True
                if float(series) > 0:
                    it_is = True
                else: 
                    it_is = False
            except ValueError:
                it_is = False
            
        series = float(series)
        it_is = False
        while it_is == False:
            start_temp = (input('What is the starting temperature? (Cannot be less than 40F, 5C, or 278K) \n'))
            try:
                float(start_temp)
                it_is = True
            except ValueError:
                it_is = False
        start_temp = float(start_temp)
        

        
        if metric == 'Fahrenheit':
            while metric == 'Fahrenheit' and start_temp < 40: 
                print('Error temperature range is below current equation capacity!')
                it_is = False
                while it_is == False:
                    start_temp = (input('What is the starting temperature? (Cannot be less than 40F) \n'))
                    try:
                        float(start_temp)
                        it_is = True
                    except ValueError:
                        it_is = False
                start_temp = float(start_temp)

            while metric == 'Fahrenheit' and series*num_solns+start_temp > 120 or start_temp < 40: 
                print('Error temperature range is beyond current equation capacity!')
                it_is = False
                while it_is == False:
                    series = (input('What is the constant value between each temperature? \n'))
                    try:
                        float(series)
                        it_is = True
                        if float(series) > 0:
                            it_is = True
                        else: 
                            it_is = False
                    except ValueError:
                        it_is = False
                series = float(series)
                it_is = False
                while it_is == False:
                    start_temp = (input('What is the starting temperature? (Cannot be less than 40F nor can the series of solutions exceed 120F) \n'))
                    try:
                        float(start_temp)
                        it_is = True
                    except ValueError:
                        it_is = False
                start_temp = float(start_temp)
                
                
                
        elif metric == 'Celsius':
            while metric == 'Celsius' and start_temp < 5: 
                print('Error temperature range is below current equation capacity!')
                it_is = False
                while it_is == False:
                    start_temp = (input('What is the starting temperature? (Cannot be less than 5C) \n'))
                    try:
                        float(start_temp)
                        it_is = True
                    except ValueError:
                        it_is = False
                start_temp = float(start_temp)
            
            while metric == 'Celsius' and series*num_solns+start_temp > 50 or start_temp < 5: 
                print('Error temperature range is beyond current equation capacity!')
                it_is = False
                while it_is == False:
                    series = (input('What is the constant value between each temperature? \n'))
                    try:
                        float(series)
                        it_is = True
                        if float(series) > 0:
                            it_is = True
                        else: 
                            it_is = False
                    except ValueError:
                        it_is = False
                series = float(series)
                it_is = False
                while it_is == False:
                    start_temp = (input('What is the starting temperature? (Cannot be less than 5C nor can the series of solutions exceed 50C) \n'))
                    try:
                        float(start_temp)
                        it_is = True
                    except ValueError:
                        it_is = False
                start_temp = float(start_temp)

                
                
        elif metric == 'Kelvin':
            while metric == 'Kelvin' and start_temp < 278: 
                print('Error temperature range is below current equation capacity!')
                it_is = False
                while it_is == False:
                    start_temp = (input('What is the starting temperature? (Cannot be less than 278K) \n'))
                    try:
                        float(start_temp)
                        it_is = True
                    except ValueError:
                        it_is = False
                start_temp = float(start_temp)
                
                '''start_temp = float(input('What is the starting temperature? (Cannot be less than 278K) \n'))'''
            
            while metric == 'Kelvin' and series*num_solns+start_temp > 328 or start_temp < 278: 
                print('Error temperature range is beyond current equation capacity!')
                it_is = False
                while it_is == False:
                    series = (input('What is the constant value between each temperature? \n'))
                    try:
                        float(series)
                        it_is = True
                        if float(series) > 0:
                            it_is = True
                        else: 
                            it_is = False
                    except ValueError:
                        it_is = False
                series = float(series)
                it_is = False
                while it_is == False:
                    start_temp = (input('What is the starting temperature? (Cannot be less than 278K nor can the series of solutions exceed 328K) \n'))
                    try:
                        float(start_temp)
                        it_is = True
                    except ValueError:
                        it_is = False
                start_temp = float(start_temp)
               
        for i in range(0,num_solns):
            if metric == 'Fahrenheit':
                IPA = (start_temp - 178.5698046)/(-2.392698462)
            elif metric == 'Celsius':
                IPA = (start_temp - 81.42766923)/(-1.329276923)
            else:
                IPA = (start_temp - 273.15 - 81.42766923)/(-1.329276923)
            vol_IPA = vol_solns * (IPA/100)
            vol_MEOH = vol_solns-vol_IPA
            list_of_IPA.append("{:.3f}".format(vol_IPA))
            list_of_MEOH.append("{:.3f}".format(vol_MEOH))
            temps.append("{:.1f}".format(start_temp))
            start_temp = start_temp+series

        print('\nRECIPE:\n| Temperature ('+metric+') | Mass CoCl2*6H2O (grams): | Volume Isopropanol (mL): | Volume Methanol (mL): |')
        print('-'*len('| Temperature ('+metric+') | Mass CoCl2*6H2O (grams): | Volume Isopropanol (mL): | Volume Methanol (mL): |'))
        for i in range(0,len(temps)):
            if metric!= 'Celsius' and metric!= 'Kelvin':
                print('|'+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+str(temps[i])+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+'|'+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+str("{:.4f}".format(mass))+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+'|'+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+str(list_of_IPA[i])+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+' |'+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+str(list_of_MEOH[i])+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+'|')
            elif metric== 'Celsius' and len(temps[i])==3 and metric!= 'Kelvin' and metric!= 'Fahrenheit':
                print('|'+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+str(temps[i])+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+'|'+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+str("{:.4f}".format(mass))+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+'|'+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+str(list_of_IPA[i])+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+' |'+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+str(list_of_MEOH[i])+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+'|')
            else:
                print('|'+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+str(temps[i])+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+' |'+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+str("{:.4f}".format(mass))+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+'|'+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+str(list_of_IPA[i])+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+' |'+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+str(list_of_MEOH[i])+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+'|')
            
            '''elif metric== 'Celsius' and len(temps[i])==3 or metric!= 'Kelvin' or metric!= 'Fahrenheit':
                print('|'+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+str(temps[i])+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+'|'+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+str("{:.4f}".format(mass))+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+'|'+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+str(list_of_IPA[i])+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+' |'+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+str(list_of_MEOH[i])+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+'|')'''
            
            '''if metric!= 'Kelvin':
                print('|     '+str(temps[i])+'      |   '+str("{:.4f}".format(mass))+'   |        '+str(list_of_IPA[i])+'        |      '+str(list_of_MEOH[i])+'      |')
            else:
                print('|     '+str(temps[i])+'     |   '+str("{:.4f}".format(mass))+'   |        '+str(list_of_IPA[i])+'        |       '+str(list_of_MEOH[i])+'      |')'''

                
                
                
    if num_solns == 1:
        it_is = False
        while it_is == False:
            temp = (input('At what temperature do you want the solution to change color? \n'))
            try:
                float(temp)
                it_is = True
                if metric == 'Fahrenheit': 
                    if float(temp) >= 40 and float(temp)<=120:
                        it_is = True
                    else:
                        print('Temperature value must be between 40F and 120F!')
                        it_is = False
                if metric == 'Celsius': 
                    if float(temp) >= 5 and float(temp)<=50:
                        it_is = True
                    else:
                        print('Temperature value must be between 5C and 50C!')
                        it_is = False
                if metric == 'Kelvin': 
                    if float(temp) >= 278 and float(temp)<=328:
                        it_is = True
                    else:
                        print('Temperature value must be between 278K and 328K!')
                        it_is = False
            except ValueError:
                it_is = False
        temp = float(temp)
                
        if metric == 'Fahrenheit':
            IPA = (temp - 178.5698046)/(-2.392698462)
        elif metric == 'Celsius':
            IPA = (temp - 81.42766923)/(-1.329276923)
        else:
            IPA = (temp-273.15- 81.42766923)/(-1.329276923)
        vol_IPA = vol_solns * (IPA/100)
        vol_MEOH = vol_solns-vol_IPA
        print('\nThe recipe for a solution that changes color at a temperature of '+str(temp)+ ' ' +str(metric)+' is: \nMass CoCl2*6H2O:      '+str("{:.4f}".format(mass))+' grams \nVolume Isopropanol:   '+ str("{:.3f}".format(vol_IPA)) + ' mL\nVolume Methanol:      ' + str("{:.3f}".format(vol_MEOH))+' mL')
        
        
        
        
    if series_yesno == 'NO' or series_yesno == 'No' or series_yesno == 'no' or series_yesno == 'n' or series_yesno == 'N':
        print('Temperature values must be between 40-120F, 5-50C, or 278-328K.')
        for i in range (0,num_solns):
            it_is = False
            while it_is == False:
                temp = (input('What is temperature #'+ str(i+1)))
                try:
                    float(temp)
                    it_is = True
                    
                    if metric == 'Fahrenheit': 
                        if float(temp) >= 40 and float(temp)<=120:
                            it_is = True
                        else:
                            print('Temperature value must be between 40F and 120F!')
                            it_is = False
                    if metric == 'Celsius': 
                        if float(temp) >= 5 and float(temp)<=50:
                            it_is = True
                        else:
                            print('Temperature value must be between 5C and 50C!')
                            it_is = False
                    if metric == 'Kelvin': 
                        if float(temp) >= 278 and float(temp)<=328:
                            it_is = True
                        else:
                            print('Temperature value must be between 278K and 328K!')
                            it_is = False
                    if "{:.1f}".format(float(temp)) in temps:
                        print('That temperature has already been entered, try a new value!')
                        it_is = False
                except ValueError:
                    it_is = False

            temp = float(temp)
            if metric == 'Fahrenheit':
                IPA = (temp - 178.5698046)/(-2.392698462)
            elif metric == 'Celsius':
                IPA = (temp - 81.42766923)/(-1.329276923)
            else:
                IPA = (temp-273.15- 81.42766923)/(-1.329276923)
            vol_IPA = vol_solns * (IPA/100)
            vol_MEOH = vol_solns-vol_IPA
            list_of_IPA.append("{:.3f}".format(vol_IPA))
            list_of_MEOH.append("{:.3f}".format(vol_MEOH))
            temps.append("{:.1f}".format(temp))
            temps.sort()
            
            

        print('\nRECIPE:\n| Temperature ('+metric+') | Mass CoCl2*6H2O (grams): | Volume Isopropanol (mL): | Volume Methanol (mL): |')
        print('-'*len('| Temperature ('+metric+') | Mass CoCl2*6H2O (grams): | Volume Isopropanol (mL): | Volume Methanol (mL): |'))
        
        for i in range(0,len(temps)):
            if metric!= 'Celsius' and metric!= 'Kelvin':
                print('|'+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+str(temps[i])+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+'|'+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+str("{:.4f}".format(mass))+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+'|'+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+str(list_of_IPA[i])+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+' |'+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+str(list_of_MEOH[i])+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+'|')
            elif metric== 'Celsius' and len(temps[i])==3 and metric!= 'Kelvin' and metric!= 'Fahrenheit':
                print('|'+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+str(temps[i])+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+'|'+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+str("{:.4f}".format(mass))+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+'|'+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+str(list_of_IPA[i])+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+' |'+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+str(list_of_MEOH[i])+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+'|')
            else:
                print('|'+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+str(temps[i])+' '*int((len(' Temperature ('+metric+') ')-len(str(temps[i])))/2)+' |'+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+str("{:.4f}".format(mass))+' '*int((len(' Mass CoCl2*6H2O (grams): ')-len(str("{:.4f}".format(mass))))/2)+'|'+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+str(list_of_IPA[i])+' '*int((len(' Volume Isopropanol (mL): ')-len(str(list_of_IPA[i])))/2)+' |'+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+str(list_of_MEOH[i])+' '*int((len(' Volume Methanol (mL): ')-len(str(list_of_MEOH[i])))/2)+'|')

            '''if metric!= 'Kelvin':
                print('|          '+str(temps[i])+'          |        '+str("{:.4f}".format(mass))+'       |           '+str(list_of_IPA[i])+'           |         '+str(list_of_MEOH[i])+'          |')
            else:
                print('|          '+str(temps[i])+'         |        '+str("{:.4f}".format(mass))+'       |           '+str(list_of_IPA[i])+'           |          '+str(list_of_MEOH[i])+'         |')'''
        
main()
```
